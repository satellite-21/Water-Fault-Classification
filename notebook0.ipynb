```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../internal dataset/Hostel-20230521T172131Z-001/Hostel/export (1).csv')
```

```python
data['Date Time'] = pd.to_datetime(data['Date Time'], format='%m/%d/%Y %H:%M')
```

```python
data.set_index('Date Time', inplace=True)
```

using sampling as sampling interval '1H'

```python
# Perform fixed interval sampling (e.g., resampling every hour)
sampling_interval = '1H'
data_resampled = data.resample(sampling_interval).mean()
```

```python
# Plot the resampled time series data using seaborn
sns.set(style='darkgrid')
plt.figure(figsize=(16, 7))
```

```python
def plot_df(df, x, y, title="", xlabel='DateTime', ylabel='Slave_Device1_CH1_FLOW m3/h', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
```

```python
data_resampled.info()
```

```python
plot_df(data_resampled, x=data_resampled.index, y=data_resampled['Slave_Device1_CH1_FLOW m3/h'], title='Slave_Device1_CH1_FLOW m3/h')
```

```python
def sampling_data_plot(df, x, field):
    sampling_interval = x
    data_resampled = data.resample(sampling_interval).mean()
    sns.set(style='darkgrid')
    plt.figure(figsize=(32, 14))
    plot_df(data_resampled, x=data_resampled.index, y=data_resampled[field], title=field)
```

```python
sampling_data_plot(data, '30T', 'Slave_Device1_CH1_FLOW m3/h')
```

```python
sampling_data_plot(data, '30T', 'Slave_Device1_CH2_VELOCITY m/s')
```

```python
sampling_data_plot(data, '30T', 'Slave_Device1_CH3_Pressure BAR')
```
